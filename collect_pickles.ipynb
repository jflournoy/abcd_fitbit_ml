```python
#!/users/jflournoy/.conda/envs/abcd_ml_3.7/bin/python
print('importing libs...')
import numpy as np
import pandas as pd
import argparse
from joblib import dump, load
import logging
from pathlib import Path

listing = [x for x in Path('out/').rglob('*.pkl')]
filesdf = pd.DataFrame([str(x.stem).split('_') for x in listing],
            columns = ['script', 'predictors', 'outcome', 'summary', 'time', 'cv_iter', 'slurmid', 'arrayid'])
filesdf['fn'] = [str(x) for x in listing]
#filesdf.drop()
filesdf.drop(filesdf[filesdf['fn'].str.contains('old', na=False)].index, inplace=True)

##Rewrite to collect into a list and then concat into a df
scores_list = []
test_data_df_list = []
for index, row in filesdf.iterrows():
    print("Reading {}".format(row.fn))
    with open(row.fn, 'rb') as f:
        out_dict = load(f)
    scores_list.append(pd.DataFrame({key: [out_dict[key]] for key in ['score_test', 'score_train']}))
    test_data_df_chunk = pd.DataFrame({key: list(np.concatenate(out_dict[key]).flat) for key in ['Y_test', 'Y_pred_test']})
    for i in row.index:
        test_data_df_chunk[i] = row[i]
    out_test_data_csv_fn = "csv/{}.csv".format(Path(row.fn).stem)
    test_data_df_list.append(test_data_df_chunk)
    print("Writing {}".format(out_test_data_csv_fn))
    test_data_df_chunk.to_csv(out_test_data_csv_fn)
```
